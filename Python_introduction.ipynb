## Basic Python data structures

```python
# Data types

x = 11              # int or integer
y = 5.5             # floats or floating point numbers
z = "Hello World!"  # string
a = True            # boolean
b = (3, 4)          # pair or tuple

print(x + 4)
```

```python
# We can turn one data type into another if it makes sense:

x = "100"
print(int(x))

x = 98.45
print(int(x))
```

```python
# Lists

L = [10, 3, 4, 5]

L = ["Hello World!", 10, 5.5, True]
#.    0.             1.   2.   3

print(L[2])

L[2] = 6.5

print(L[2])
```

```python
# For loops

for i in range(25): # range(25) means numbers from 0 to (25 - 1)
    print(i)
```

```python
L = [3, 1, 4, 2, 84, 3, 12]

# we can go thorugh elements in L by index
sum = 0
for i in range(7):
  sum += L[i]

print(sum)

# we can go thorugh elements in L by "for number in L:"

sum = 0
for number in L:
  sum += number

print(sum)
```

```python
# We can add lists to another list to represent a matrix as follows

# We intialize the matrix as empty (10x10)
M = []

for i in range(10):

    # We create a list for the i-th row
    R = []
    for j in range(10):
        R.append(i)

    # Add the new row to the matrix
    M.append(R)

M
```

```python
# Can access a specific coordinate as follow

M[7][3]
```

```python
# Split operation for strings

s = "This is the Python introduction for IEOR4004"
print(s.split())
print(s.split()[4])
```

```python
# Sets

S = {10, 3, 3, 3, 4}
# S = set([10, 3, 3, 3, 4])

print(S)

print(5 in S)

S.add(5)

print(5 in S)

S.remove(3)

print(3 in S)
```

```python
# Dictionary

D = {"Corn": 3, "Wheat": 7}

print(D["Corn"])

D["Rice"] = 5
D["Corn"] = 4

print(D)
```

## File Management

```python
with open("file.txt", "w") as file:
  file.write("Hello 1\n")
  file.write("Hello 2\n")
  file.write("Hello 3\n")
```
